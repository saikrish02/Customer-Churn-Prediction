```python
import pandas as pd
df = pd.read_csv('churn_prediction.csv')
df
```

```python
df1 = df[df.TotalCharges != ' ']
df1
```

```python
df1.TotalCharges = pd.to_numeric(df1.TotalCharges)

df1.dtypes
```

```python
tenure_no = df1[df1.Churn == 'No'].tenure 
tenure_yes = df1[df1.Churn == 'Yes'].tenure

from matplotlib import pyplot as plt 

plt.hist([tenure_yes, tenure_no], color=['green', 'red'])
```

```python
for columns in df1:
  if df1[columns].dtype == 'object':
    print(f'{columns} {df1[columns].unique()}')
```

```python
df1.replace('No internet service', 'No', inplace=True)
df1.replace('No phone service', 'No', inplace=True)
```

```python
df1.drop(columns='customerID', inplace=True)
```

```python
columns_yes_no = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']
for column in columns_yes_no:
  df1[column].replace({'Yes': 1, 'No':0}, inplace=True)
```

```python
df1['gender'].replace({'Male':0, 'Female':1}, inplace=True)
df1.dtypes
```

```python
df2 = pd.get_dummies(data=df1, columns=['PaymentMethod', 'Contract', 'InternetService'])
df2.dtypes
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()


columns_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

df2[columns_to_scale] = scaler.fit_transform(df2[columns_to_scale])
```

```python
df2.sample(5)
```

```python
x = df2.drop(columns='Churn')
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=5)

y_test
```

```python
import tensorflow as tf
from tensorflow import keras 
```

```python
model = keras.Sequential([
    keras.layers.Dense(26, input_shape=(26,),activation='relu'),
    keras.layers.Dense(1, activation='relu')
])

model.compile(
    optimizer='adam', 
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs=500)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_predicted = model.predict(X_test)

y_pred = []

for i in y_predicted:
  if i > 0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)

y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
